```python
import matplotlib.pyplot as plt

# Sample COVID-19 data
labels = ["Active Cases", "Recovered", "Deaths"]
cases = [500000, 1200000, 30000]  # Example numbers

# Define colors
colors = ["skyblue", "lightgreen", "red"]

# Create pie chart
plt.figure(figsize=(7, 7))
plt.pie(cases, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})

# Add title
plt.title("COVID-19 Case Distribution")

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample COVID-19 data (in millions)
countries = ["USA", "India", "Brazil", "Russia", "UK"]
active_cases = [2.5, 1.8, 1.2, 0.9, 0.7]
recovered = [28.5, 30.2, 19.8, 8.5, 7.9]
deaths = [0.8, 0.4, 0.6, 0.3, 0.2]

# Define bar positions
x = np.arange(len(countries))

# Plot stacked bars
plt.figure(figsize=(8, 6))
plt.bar(x, active_cases, label="Active Cases", color="skyblue")
plt.bar(x, recovered, bottom=active_cases, label="Recovered", color="lightgreen")
plt.bar(x, deaths, bottom=np.array(active_cases) + np.array(recovered), label="Deaths", color="red")

# Labels and title
plt.xticks(x, countries)
plt.xlabel("Countries")
plt.ylabel("Number of Cases (in millions)")
plt.title("COVID-19 Case Distribution by Country")
plt.legend()

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample COVID-19 data
data = {
    "Country": ["USA", "India", "Brazil", "Russia", "UK", "France", "Italy", "Germany"],
    "Total Cases": [50000000, 45000000, 35000000, 25000000, 15000000, 18000000, 16000000, 14000000],
    "Deaths": [800000, 530000, 620000, 350000, 230000, 280000, 220000, 180000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set Seaborn style
sns.set_style("whitegrid")

# Create scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Total Cases", y="Deaths", hue="Country", size="Total Cases", sizes=(50, 300), palette="coolwarm", edgecolor="black")

# Labels and title
plt.xlabel("Total COVID-19 Cases")
plt.ylabel("Total Deaths")
plt.title("COVID-19 Cases vs Deaths by Country")

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample COVID-19 data for Italy (Replace with real data)
dates = pd.date_range(start="2022-01-01", periods=10, freq="M")
data = {
    "Date": dates,
    "Active Cases": [500000, 450000, 400000, 350000, 300000, 250000, 200000, 150000, 100000, 50000],
    "Recovered": [1000000, 1100000, 1200000, 1300000, 1400000, 1500000, 1600000, 1700000, 1800000, 1900000],
    "Deaths": [140000, 145000, 150000, 155000, 160000, 165000, 170000, 175000, 180000, 185000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Compute correlation matrix
corr = df.drop(columns=["Date"]).corr()

# Set Seaborn style
sns.set_style("whitegrid")

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")

# Labels and title
plt.title("COVID-19 Correlation Heatmap - Italy")

# Show plot
plt.show()
```
